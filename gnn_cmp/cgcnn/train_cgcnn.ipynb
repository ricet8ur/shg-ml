```python
# .json -> id_prop.csv + *.cifs inside ./data/regression/
def prepare_data_for_regression(ds_name="dataset_of_max_abs_shg"):
    total_ds = load_from_json(f"../../data/final_data/{ds_name}.json")
    data_regression = "./data/regression/"

    try:
        ! rm -r ./data/regression/
    except Exception as e:
        print(e)
    ! mkdir -p ./data/regression
    try:
        os.mkdir("trained")
    except:
        pass
    try:
        os.mkdir("trained/" + ds_name)
    except:
        pass
    try:
        os.mkdir(data_regression + ds_name)
    except:
        pass
    shutil.copy("./data/sample-regression/atom_init.json", data_regression + ds_name)
    names, labels = [], []

    for k, v in total_ds.items():
        structure: Structure = Structure.from_dict(v["structure"])
        structure.relabel_sites(True).to_file(
            data_regression + ds_name + "/" + k + ".cif", "cif"
        )
        names.append(k)
        labels.append(v["shg"])
    print(len(names))
    pd.DataFrame.from_dict(dict(names=names, labels=labels)).to_csv(
        data_regression + ds_name + "/id_prop.csv", header=False, index=False, sep=","
    )
```

```python
def finetune(ds_name:str='dataset_of_max_abs_shg'):
    prepare_data_for_regression(ds_name)

    ds_path = "./data/regression/"
    sys.argv = [ds_path, ds_path]
    from main import args, main
    args.disable_cuda = True
    args.batch_size = 32
    args.epochs = 530
    # args.n_h = 2
    args.h_fea_len = 32
    args.n_conv = 4
    args.data_options = [ds_path]
    args.lr_milestones = [500]
    args.resume = "./pre-trained/band-gap.pth.tar"


    from aim import Run

    # Initialize a new run
    run = Run(
        experiment=f"CGCNN FT from band-gap.pth.tar on {ds_name}",
        log_system_params=True,
    )

    # Log run parameters
    run["hparams"] = vars(args)
    main(run, ds_name)
    run.close()
```
