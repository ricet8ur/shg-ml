```python
import numpy as np
import orjson
import sys
sys.path.append("../../")
from utils.save_and_load import load_from_json, save_to_json
```

```python
data = load_from_json('./artifacts/c8246bf979534b69a11ae55b/train_info.json')
```

```python
crystal_pred = data['crystal_predictions_log_for_all_folds']['fold']['0']
targets = data['dataset_index_key_target']
```

```python
targets
```

```python
pred, tg = [], []
for k, v in crystal_pred.items():
    print(targets[k], v[-1])
    tg.append(targets[k]['target'])
    pred.append(v[-1])

from sklearn import metrics
from scipy import stats

mae = metrics.mean_absolute_error(tg, pred)
rmse = metrics.root_mean_squared_error(tg, pred)
r2 = metrics.r2_score(tg, pred)
spearmanr = stats.spearmanr(tg, pred)
test_metrics = dict(
    mae=mae,
    rmse=rmse,
    r2=r2,
    spearmanr_statistic=spearmanr.statistic,
    spearmanr_pvalue=spearmanr.pvalue,
)
test_metrics
```
