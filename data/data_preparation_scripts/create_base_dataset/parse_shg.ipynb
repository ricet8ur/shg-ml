```python
# str_xi to dict of d_ijk

import itertools

def transform_string_to_dict(input_string):
    key_value = input_string.split('=')
    
    key = key_value[0][1:]
    value = float(key_value[1].strip())
    
    permutations = set(''.join(p) for p in itertools.permutations(key))
    
    result_dict = {perm: value for perm in permutations}
    
    return result_dict

input_string = "a123=124.23"
result = transform_string_to_dict(input_string)
print(result)
```

```python
def generate_combinations(characters, length):
    combinations = itertools.product(characters, repeat=length)
    
    result = [''.join(combination) for combination in combinations]
    
    return result

characters = ['1', '2', '3']
length = 3
combinations = generate_combinations(characters, length)

for combo in combinations:
    print(combo)
```
